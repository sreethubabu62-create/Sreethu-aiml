```python
def expenses_made():
  expenses=[]
  number_of_days=int(input("Enter the number of days"))
  for i in range(number_of_days):
    day_name=input(f"Enter the day name")

    seed_cost=int(input(f"Enter the seed cost:"))
    fertilizer_cost=int(input(f"Enter fertilizer cost:"))
    labour_cost=int(input("Enter the labour cost:"))

    expenses.append([day_name,[seed_cost,fertilizer_cost,labour_cost]])
  return expenses

total_expenditure=expenses_made()
print("Daily expenses")
for expenditure in total_expenditure:
  print(expenditure[0],"expenses are",expenditure[1])
```

```python
"""Tuple length:To determine how many items a tuple has,use the len()function"""

thistuple=("apple",)
print(len(thistuple))
```

```python
#Create tuple with one item
"""To create a tuple with only one item,you have to add a comma after the item,otherwise Python will not recognize it as a tuple"""

thistuple=("apple",)
print(type(thistuple))

#Not a tuple
thistuple=("apple")
print(type(thistuple))
```
