```python
a=10
print("Initial value:",a)

a+=5
print("After+=",a)

a-=3
print("After-=",a)

a*=2
print("After*=",a)

a/=4
print("After/=",a)

a//=2
print("After//=",a)

a%=3
print("After%=",a)

a**=2
print("After**=",a)
```

```python
x=20
y=15

print("x==y : ",x==y)
print("x!=y : ",x!=y)
print("x>y : ",x>y)
print("x<y : ",x<y)
print("x>=y : ",x>=y)
print("x<=y : ",x<=y)
```

```python
rainfall = True
soil_moisture = False

print("AND : ",rainfall and soil_moisture)
print("OR : ",rainfall or soil_moisture)
print("NOt : ",not rainfall)
```

```python
a=10
b=6

print("Bitwise AND(&) : ",a & b)
print("Bitwise OR(|) : ",a | b)
print("Bitwise XOR(^) : ",a ^ b)
print("Bitwise NOT(~a) : ",~a)
print("Left Shift(a<<1) : ",a << 1)
print("Right Shift(a>>1) : ",a >> 1)
```
