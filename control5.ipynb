```python
#Accept soil moisture(%) and rainfall(mm)
moisture = float(input("Enter soil moisture(%): "))
rainfall = float(input("Enter rainfall(mm): "))

#Check suitability for Rice
if moisture >= 70 and rainfall >=200:
  print("Suitable for Rice")
else:
  print("Not suitable for Rice")
```

```python
#Accept crop name and soil nitrogen level
crop=input("Enter crop name(Rice,Wheat,Maize): ")
nitrogen=float(input("Enter soil nitrogen level: "))

#Fertilizer recommendation
if crop=="Rice" and nitrogen<50:
  print("Apply Urea")
elif crop=="Wheat" and nitrogen<40:
  print("Apply DAP")
else:
  print("Fertilizer not required")
```

```python
#Accept rainfall values for 5 days
rainfall=[float(input(f"Enter rainfall for day {i+1}: "))for i in range(5)]

#Calculate average rainfall
avg_rainfall=sum(rainfall)/ len(rainfall)
print(f"Average rainfall:{avg_rainfall}")

#Display result
if avg_rainfall>=100:
  print("Sufficient rainfall")
else:
  print("Insufficient rainfall")
```
