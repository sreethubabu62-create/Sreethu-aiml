```python
crop_list = ["Rice", "Wheat", "Maize"]

crop_name = input("Enter crop name: ").strip().capitalize()

if crop_name in crop_list:
    print("Crop available")
else:
    print("Crop not available")
```

```python
rainfall = []

for i in range(5):
    value = float(input(f"Enter rainfall for day {i+1}: "))
    rainfall.append(value)

average = sum(rainfall) / 5
print("Average rainfall:", average)

if average >= 100:
    print("Sufficient rainfall")
else:
    print("Insufficient rainfall")
```
