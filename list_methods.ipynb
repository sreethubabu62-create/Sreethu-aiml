```python
#Negative Indexing
thislist=["apple","banana","cherry"]
print(thislist[-2])
```

```python
#Range of indexes
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[2:4])
```

```python
#This example returns the items from the beginning to,but NOT including
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[2:])
```

```python
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[-4:-1])
```
