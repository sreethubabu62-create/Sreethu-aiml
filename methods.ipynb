```python
#type()
#From Python's Perspective,lists are defined as objects with the data type
x=10
y=11.5
mylist=["Apple","Banana","Cherry"]

print(type(x))
print(type(y))
print(type(mylist))
```

```python
#index()method
#Returns the index of the first occurence of an element
thislist=["apple","banana","cherry","banana"]

index_value=thislist.index("banana")
print("Index of'banana':",index_value)
```

```python
#count()method
#count how many times an element appears in the list

thislist=["apple","banana","apple","cherry","apple"]

count_value=thislist.count("apple")
print("Count of'apple':",count_value)
```

```python
#reverse()method
#Reverses the order of elements in the list(in-place)

thislist=["orange","mango","kiwi","banana"]
print("Original list:",thislist)
thislist.reverse()
print("Reversed list:",thislist)
```

```python
original_list=[1,2,3,4,5]
reversed_iterator=reversed(original_list)
reversed_list=list(reversed_iterator) #convert the iterator to a list
print(reversed_list)
#Output:[5,4,3,2,1]
```

```python
#List Length
#To dtermine how many items a list has,use the len()function:

thislist=["apple","banana","cherry"]
print(len(thislist))
```

```python
#+list Items-Data types

#List items can be of any data type:

list1=["apple","banana","cherry"]
list2=[1,5,7,9,3]
list3=[True,False,False]

print(list1)
print(list2)
print(list3)
```

```python
# Using len to access elements,but should be cautions while using it
for item in range(len(list1)):
  print(item)
```
