```python
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
clean_codes=[]

for code in codes:
   if code.startswith("CRP_"):
    clean_codes.append(code[4:])
print(clean_codes)
```

```python
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]

for crop_code in codes:
   if crop_code.startswith("CRP_"):
    code,crop,year=crop_code.split("_")
    print(crop,year)
```
