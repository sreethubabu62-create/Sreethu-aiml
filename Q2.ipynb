```python
 crops=[25,18,35] #yields in tons/hectare
 pairs=[]

 for i in range(len(crops)):
   for j in range(i+1,len(crops)):
     if crops[i]+crops[j]>50:
      pairs.append((crops[i],crops[j],"High"))
     elif crops[i]+crops[j]>40:
      pairs.append((crops[i],crops[j],"Medium"))
 print(pairs)
```

```python
fields=[[10,0,15],[0,20,0],[5,10,25]]
# soil moisture
irrigated=0

for row in range(len(fields)):
  for col in range(len(fields[row])):
    if fields[row][col]<12:
      irrigated += 1
      continue
    if fields[row][col]>20:
      break
  print(irrigated)
```

```python
# Harvest Window Calculation

days=[5,3,7,2] # days to harvest
windows=[]
i=0
while i<len(days)-1:
  j=i+1
  while j<len(days):
    if days[i]+days[j]<=8:
      windows.append((i,j))
    j +=1
  i+=1
print(windows)
```

```python
nums=[1,2,3]
output=[]

for i in range(len(nums)):
  for j in range(len(nums)):
    if i == j:
      continue
    if nums[i]+nums[j]>4:
      output.append((nums[i],nums[j]))
print(output)
```

```python
def filter_positive(lst):
  result=[]
  for num in lst:
    if num>0:
      result.append(num*2)
  return result
filter_positive([1,2,-3,4])
```

```python
def test_control(tup):
  count_even=0
  for val in tup:
   if val%2==0:
     count_even+=1
   elif val>2:
     continue
  return count_even
test_control((1,2,3,4))
```
