```python
def water_needs(plant_type):
  if plant_type =="corn":
   return 30 #corn needs 30 litres of water per day
  elif plant_type =="wheat":
   return 20 #wheat needs 20 litres of water per day
  else:
    return 25 #default water needs
water_needs("wheat")
```

```python
#Function for addition
def add(x,y):
  return x+y

#Function for subtraction
def subtract(x,y):
  return x-y

#Function for multiplication
def multiply(x,y):
  return x*y

#Function for divison
def divide(x,y):
  if y==0:
    return"Error!division by zero."
  else:
    return x/y

#Function to take input values from the user
def get_input():
  try:
    #Take users input for the first number
    x=float(input("Enter the first number:"))
    #Take users input for the second number
    y=float(input("Enter the second number:"))
    return x,y
  except ValueError:
    print("Invalid input! Please enter valid numbers.")
    return get_input() # Recursive call to get valid input

#Function to print the menu of operations
def print_menu():
  print("\nSelect operations.")
  print("1.Add")
  print("2.Subtract")
  print("3.Multiply")
  print("4.Divide")
  print("5.Exit")

#Main calculator function
def calculator():
  while True:
    print_menu()

    #Get the user's choice
    choice=input("Enter choice(1/2/3/4/5):")
    if choice=='5':
     print("Exiting the calculator.")
     break #Exit the program

   #Call get_input function to get the values
    x,y=get_input()

   #Perform operation based on user choice
    if choice=='1':
     print(f"{x}+{y}={add(x,y)}")
    elif choice=='2':
     print(f"{x}-{y}={subtract(x,y)}")
    elif choice=='3':
     print(f"{x}*{y}={multiply(x,y)}")
    elif choice=='4':
     print(f"{x}/{y}={divide(x,y)}")
    else:
     print("Invalid Input!Please select a valid operation(1/2/3/4/5).")

#Directly call the calculator function
calculator()
print("done with program")

```
