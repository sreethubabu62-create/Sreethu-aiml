```python
import numpy as np
#Crop yields for 4 farms over 5 years(tons/hectare)
yields=np.array([
    [45,48,52,49,53],
    [38,41,39,44,42],
    [52,55,58,56,60],
    [41,43,45,47,46]
])

#Weather impact factors for each year(1D-array_broadcasts across farms)
weather_factors=np.array([1.0,0.95,1.05,0.98,1.02])

print('Original Yields(tons/hectare):')
print(yields)
print('\nWeather Impact Factors:',weather_factors)

#Broadcasting:adjust all farm yields by weather factors
adjusted_yields=yields*weather_factors
print('\nWeather-Adjusted Yields:')
print(adjusted_yields.round(2))

#Boolean indexing:find exceptional years
exceptional=adjusted_yields>55
print('\nExceptional Years(>55):')
print(adjusted_yields[exceptional])

#Conditional operations:classify yields
categories=np.where(
    adjusted_yields>55,'High',
    np.where(adjusted_yields>40,'Medium','Low')
)
print('\nYield categories:')
print(categories)

#Advanced indexing:select specific farms
selected_farms=[0,2] #Farm 1 and Farm 3
print(f'\nSelected farm yields:')
print(yields[selected_farms])

#Multi-condition filtering
good_years=adjusted_yields[(adjusted_yields>45)&(adjusted_yields<55)]
print('\nGood performing years(45-55):')
print(good_years)
```
