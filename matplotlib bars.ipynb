```python
import numpy as np
import matplotlib.pyplot as plt

#Generate random data
data=np.random.normal(170,10,250)

#Create histogram
plt.hist(data,bins=20,color="silver")

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
data=[
    30,35,40,42,45,47,48,49,
    50,50,51,52,53,55,57,60,
    62,65,70,75
]
plt.figure(figsize=(8,5))
plt.hist(
    data,bins=5,color="pink",
    edgecolor="black",
    alpha=0.5,density=True,
    range=(30,80)
)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram without random data")
plt.grid(True)
plt.tightlayout()
plt.show()
```
