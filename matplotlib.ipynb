```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,25,20,30,40]
plt.plot(x,y,marker='*',
linestyle='--',color='purple',
linewidth=4,markersize=20,
label='sales')

plt.title("Monthly sales trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,15,25,30]
sizes=[50,100,200,300,150]
colors=['red','blue','green','pink','yellow']
plt.scatter(x,y,s=sizes,c=colors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.rand(50)
y=np.random.rand(50)
sizes=1000*np.random.rand(50)
colors=np.random.rand(50)

plt.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.6)
plt.title("Advanced Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
