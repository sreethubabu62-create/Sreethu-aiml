```python
import pandas as pd

data={
    "calories":[420,380,390],
    "duration":[50,40,45]
}
#load data into a DataFrame objects,
df=pd.DataFrame(data)
print(df)
```

```python
#Use a list of indexes:
print(df.loc[[0,1]])
```

```python
import pandas as pd
data={
    "calories":[420,380,390],
    "duration":[50,40,45]
}
df=pd.DataFrame(data,index=["day1","day2","day3"])
print(df)
```

```python
print(df.loc["day2"])
```

```python
import pandas as pd

data={
  "Name":["Alice","Bob","Charlie"],
  "Age":[20,22,21],
  "Marks":[85,90,88]
}
df=pd.DataFrame(data)
print(df)
```

```python
import pandas as pd

data=[
  ["Alice",20,85],
  ["Bob",22,90],
  ["Charlie",21,88]

]
df=pd.DataFrame(data,columns=["Name","Age","Marks"])
print(df)
```
