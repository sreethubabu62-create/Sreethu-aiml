```python
 rainfall=float(input("Enter rainfall in mm: "))
 if rainfall<500:
  print("Suitable crop:Millet")

elif rainfall<1000:
  print("Suitable crop:Rice")

else:
  print("Suitable crop:sugarcane")
```
